# Problem 9, Koszul connection on $T^*M$

**Goal.** A Poisson bivector $\pi$ on $M$ induces an *anchor*
$\rho := \pi^{\sharp} : T^*M \to TM$ and a *Koszul bracket* on
$1$-forms,

$$ [\alpha, \beta]_K = L_{\rho\alpha}\beta - L_{\rho\beta}\alpha - d\langle \rho\alpha, \beta\rangle. $$

Define an affine connection $\widetilde{\nabla}$ on $T^*M$ whose
function action is anchor-routed,

$$ \widetilde{\nabla}_\omega f := \rho(\omega)(f), $$

and whose vector bracket is the Koszul bracket. The Q9 chapter asks
for three families of mechanised proofs about $\widetilde{\nabla}$ on
$T^*M$, and in this notebook we close every one of them in its own
section, with the mathematical statement, the engine residue, and the
full step trace.

* **Q7 analogues**, Bianchi I & II for $\widetilde{T}, \widetilde{R}$,
  routed through the Koszul bracket $[\cdot, \cdot]_K$ rather than the
  Lie bracket on $TM$.
* **Q8 Part A analogues**, $\widetilde{\omega}^a{}_b, \widetilde{Q}_{ab}$
  are 1-forms; $\widetilde{T}^a, \widetilde{R}^a{}_b$ are 2-forms (14
  form-property props). The $\widetilde{Q}_{ab}$ non-metricity proofs
  require a cotangent-bundle metric $\widetilde{g}$ on $T^*M$, which we
  attach to the umbrella via the new ``metric=`` kwarg.
* **Q8 Part B analogues**, Cartan I & II structure equations on
  $T^*M$ with the anchor-pulled exterior derivative $\widetilde{d}$.


## Setup

A single :class:`KoszulConnectionProblem` bundles the three
sub-problems behind one facade. Internally the Bianchi facet swaps in
the `BracketApply` closure family, the form-property facet inherits
the standard rules, and the Cartan-structure facet swaps
`ExteriorDIntrinsicDefinition` for `KoszulExteriorDIntrinsicDefinition`.

```python
from jacopy.algebra.derivation import Derivation
from jacopy.brackets.koszul import KoszulBracket
from jacopy.calculus.anchor import Anchor
from jacopy.calculus.connection import koszul_connection
from jacopy.calculus.local_frame import local_frame
from jacopy.calculus.metric import metric as metric_factory
from jacopy.core.expr import Symbol
from jacopy.display.jupyter import display_chain
from jacopy.library.koszul_connection_problem import KoszulConnectionProblem
from jacopy.proof.chain import ProofChain

anchor = Anchor(name="ŌĀ")
bracket = KoszulBracket(anchor)
nabla = koszul_connection("‚ąáŐÉ", anchor=anchor, bracket=bracket)
frame = local_frame("FŐÉ", dim=3)
g = metric_factory("gŐÉ")  # cotangent-bundle metric on T*M (used only for the QŐÉ facet).

P = KoszulConnectionProblem(nabla, frame, metric=g)
P.name
```

# Q7 analogues, Bianchi identities (2 proofs)

Both identities run through `P.bianchi`, which is a `BianchiProblem`
whose closure family auto-switches to the `BracketApply` axiom set
because the connection carries a Koszul bracket (Stage 9.C). The
torsion / curvature definitions emit `BracketApply(bracket, ¬∑, ¬∑)`
payloads instead of `LieBracketVF`, and the closure family closes the
cyclic-triple Jacobi cancellation with the Koszul bracket in place.

## Q7-1, first Bianchi identity

$$
\sum_{\text{cyc}}\widetilde{R}(\alpha, \beta)\,\gamma
\;=\;
\sum_{\text{cyc}}\Bigl[(\widetilde{\nabla}_\alpha \widetilde{T})(\beta, \gamma)
\;+\; \widetilde{T}\bigl(\widetilde{T}(\alpha, \beta), \gamma\bigr)\Bigr].
$$

```python
U, V, W = Symbol("U"), Symbol("V"), Symbol("W")
res = P.bianchi.prove_first_bianchi(U, V, W)
# BianchiProblem reduces (LHS - RHS) to Zero in a single residue chain;
# rhs_final is literally Zero with no steps, so we only display the
# residue reduction (`lhs_steps`).
print("OK?", res.ok)
print("residue final:", res.lhs_final._repr_inner())
print("engine steps:", len(res.lhs_steps))
display_chain(ProofChain(res.lhs_steps))
```

```python
for i, s in enumerate(res.lhs_steps, 1):
    print(f"[{i:2}] {s.rule}")
    print(f"     {s.before._repr_inner()}")
    print(f"  ‚Üí  {s.after._repr_inner()}")
    print()
```

## Q7-2, second Bianchi identity

$$
\sum_{\text{cyc}}\Bigl[(\widetilde{\nabla}_\alpha \widetilde{R})(\beta, \gamma)\,\delta
\;+\; \widetilde{R}\bigl([\alpha, \beta]_K,\, \gamma\bigr)\,\delta\Bigr] \;=\; 0.
$$

```python
Z = Symbol("Z")
res = P.bianchi.prove_second_bianchi(U, V, W, Z)
print("OK?", res.ok)
print("residue final:", res.lhs_final._repr_inner())
print("engine steps:", len(res.lhs_steps))
display_chain(ProofChain(res.lhs_steps))
```

```python
for i, s in enumerate(res.lhs_steps, 1):
    print(f"[{i:2}] {s.rule}")
    print(f"     {s.before._repr_inner()}")
    print(f"  ‚Üí  {s.after._repr_inner()}")
    print()
```

# Q8 Part B analogues, Cartan structure equations (2 proofs)

The Cartan structure equations on $T^*M$ close through
`P.cartan_structure`, whose engine bundle swaps
`ExteriorDIntrinsicDefinition` for the anchor-pulled
`KoszulExteriorDIntrinsicDefinition` (Stage 9.E). The latter routes
its function action through `connection.function_action` (so
$\widetilde{d}$ emits $\rho(\alpha)(\cdot)$ shapes that line up with
the LHS's Y-Leibniz expansion) and emits
`BracketApply(connection.bracket, ‚Ä¶)` instead of `LieBracketVF` (so
the LHS's torsion / curvature opens to $[\alpha, \beta]_K$).

$$
