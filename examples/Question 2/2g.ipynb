# 2g, Poisson Jacobi $\Leftrightarrow [\pi,\pi]_{SN}=0$

## Problem

$$
\sum_{\mathrm{cyc}}\{f,\{g,h\}_\pi\}_\pi \;=\; 0
\qquad\Longleftrightarrow\qquad [\pi,\pi]_{SN}\;=\;0.
$$

## Strategy: definitional expansion (Vaisman / Marsden-Ratiu)

Instead of citing the **Derived Bracket Theorem** in a single
`provenance_tag="theorem"` step (the seeded-theorem shortcut, used by
`PoissonBracket.prove_jacobi_reduction`), here we drive the chain by
*definitional unfolding alone*: three definitions of the underlying
objects, each applied as one `provenance_tag="axiom"` step, take the
cyclic Poisson Jacobi sum to $\tfrac12 [\pi,\pi]_{SN}(df, dg, dh)$.

The three definitions involved:

1. **Hamiltonian-VF view of the Poisson bracket**: $\{f, X\}_\pi = X_f(X)$.
2. **Bivector view of the Poisson bracket**: $\{\varphi, \psi\}_\pi = \pi(d\varphi, d\psi)$.
3. **SN-on-bivector formula**: $\tfrac12 [\pi, \pi]_{SN}(df, dg, dh) = \sum_{\mathrm{cyc}} X_f\bigl(\pi(dg, dh)\bigr)$.

Each rewrite is encapsulated by `PoissonBracket.prove_jacobi_by_definitions`,
which returns the resulting `ProofChain`. The notebook below just calls
that method, displays the chain, and reads off the steps; the chain is
machine-built by the method, not hand-constructed.

```python
try:
    import jacopy  # noqa: F401
except ModuleNotFoundError:
    import sys
    from pathlib import Path
    here = Path.cwd().resolve()
    for candidate in (here, *here.parents):
        if (candidate / 'jacopy' / '__init__.py').is_file():
            sys.path.insert(0, str(candidate))
            break
    import jacopy  # noqa: F401

from jacopy.core.registry import PropertyRegistry
from jacopy.library.declarations import Bivector, Functions
from jacopy.library.poisson import PoissonBracket
from jacopy.library.theorem_book import theorem_book
from jacopy.display.jupyter import display_chain
```

## 1. Setup

A Poisson bivector $\pi$ and three generic functions $f, g, h$, registered
in a fresh `PropertyRegistry`. `PoissonBracket` wraps $\pi$.

```python
reg = PropertyRegistry()
pi = Bivector("pi", registry=reg)
f, g, h = Functions("f g h", registry=reg)

P = PoissonBracket(pi)
P
```

## 2. The proof chain

A single call into `PoissonBracket.prove_jacobi_by_definitions` returns
the three-step axiom-tagged `ProofChain`.

```python
chain = P.prove_jacobi_by_definitions(f, g, h, registry=reg)

print(f"steps    : {len(chain)}")
print(f"initial  : {chain.initial}")
print(f"final    : {chain.final}")
```

## 3. LaTeX render of the chain

The same chain rendered as LaTeX `gather*` block; each step shows its
`before → after` transition with the rule label.

```python
display_chain(chain)
```

## 4. Step-by-step inspection

Reading off each step explicitly: the rule applied, the justification
text the method recorded, and the `before`/`after` expressions.

```python
for i, step in enumerate(chain.steps, start=1):
    print(f"--- step {i}: rule={step.rule}, provenance={step.provenance_tag} ---")
    print(f"   justification: {step.justification}")
    print(f"   before:        {step.before}")
    print(f"   after:         {step.after}")
    print()
```

## 5. Closure check — both sides land at the same Hamiltonian form

Step 1 takes the LHS to the common Hamiltonian form $X_f(\{g,h\}_\pi) +
\text{cyclic}$. Steps 2 and 3 take the same form forward to
$\tfrac12 [\pi,\pi]_{SN}(df, dg, dh)$. Step 1's `after` and Step 2's
`before` are literally the same `Expr`, witnessing the identity at the
common form — verified structurally:

```python
step1, step2, step3 = chain.steps
print(f"step 1 after == step 2 before:  {step1.after == step2.before}")
print(f"step 2 after == step 3 before:  {step2.after == step3.before}")
print(f"chain.initial == LHS compact:   ✓ (by construction)")
print(f"chain.final == ½[π,π]_SN(df,dg,dh):  ✓ (by construction)")
```

## 6. From $\tfrac{1}{2}[\pi,\pi]_{SN}(df, dg, dh) = 0$ to $[\pi,\pi]_{SN} = 0$ — by hand

The chain above shows, by definitional rewriting alone, that

$$
\sum_{\mathrm{cyc}}\{f, \{g, h\}_\pi\}_\pi \;=\; \tfrac{1}{2}[\pi, \pi]_{SN}(df, dg, dh)
\qquad \text{for all } f, g, h \in C^\infty(M).
$$

Hence

$$
\text{Poisson Jacobi holds} \;\Longleftrightarrow\; [\pi, \pi]_{SN}(df, dg, dh) = 0
\;\;\forall f, g, h \in C^\infty(M).
$$

The map $(\alpha, \beta, \gamma) \mapsto [\pi, \pi]_{SN}(\alpha, \beta, \gamma)$
is a $C^\infty(M)$-multilinear, alternating tensor, locally every 1-form is a
$C^\infty(M)$-combination of exact forms $df_i$, so

$$
[\pi, \pi]_{SN}(df, dg, dh) = 0 \;\;\forall f, g, h
\quad\Longleftrightarrow\quad [\pi, \pi]_{SN} = 0.
$$

Combining:

$$
\boxed{\;\sum_{\mathrm{cyc}}\{f, \{g, h\}_\pi\}_\pi = 0 \;\Longleftrightarrow\; [\pi, \pi]_{SN} = 0.\;}
$$

The two `iff` arrows split into a *symbolic* part (the chain above, machine-built
from definitional axioms by `prove_jacobi_by_definitions`) and a *meta* part
(the multilinearity / locality argument, written here as commentary).

## 7. Compare with the theorem-book shortcut

The same identity is also available as a single-step theorem citation
through `prove_jacobi_reduction`, which references the seeded
`poisson_jacobi` theorem in `theorem_book`. Both routes produce the
same final identity; they differ in granularity and provenance.

```python
shortcut = P.prove_jacobi_reduction(f, g, h, registry=reg)
print(f"shortcut chain length: {len(shortcut)}")
shortcut_step, = shortcut.steps
print(f"  rule          : {shortcut_step.rule}")
print(f"  provenance    : {shortcut_step.provenance_tag}")
print(f"  justification : {shortcut_step.justification}")
print()
thm = theorem_book.get("poisson_jacobi")
print(f"theorem name      : {thm.name}")
print(f"theorem statement : {thm.statement}")
```

**Two routes to the same identity**:

| | `prove_jacobi_reduction` | `prove_jacobi_by_definitions` |
|---|---|---|
| Steps | 1 | 3 |
| `provenance_tag` | `"theorem"` | `"axiom"` × 3 |
| Cites | Derived Bracket Theorem (seeded) | three definitions only |
| Style | Paper-grade citation | Textbook-style derivation |

The package supports both because each is appropriate in a different
context: paper-style citation when invoking a known result, step-by-step
exposition when teaching or auditing the derivation.
